## Access and setup Iris data

```python
import os
import pandas as pd
import numpy as np

s = os.path.join('https://archive.ics.uci.edu', 'ml', 'machine-learning-databases', 'iris', 'iris.data')
df = pd.read_csv(s, header=None, encoding='utf-8')

y = df.iloc[0:100, 4].values
y = np.where(y == 'Iris-setosa', -1, 1)
X = df.iloc[0:100, [0, 2]].values
```

## Plot the cost corresponds to epoch

```python
import matplotlib.pyplot as plt
from classify.classifier import AdalineGD

# learn the data with ADALINE model
ada1 = AdalineGD(eta=0.01, n_iter=10).fit(X, y)
ada2 = AdalineGD(eta=0.0001, n_iter=10).fit(X, y)

# define the figure properties
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

ax[0].plot(range(1, len(ada1.cost_)+1), np.log10(ada1.cost_), marker='o')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('log(Sum-squared-error)')
ax[0].set_title('Adaline - Learning rate 0.01')

ax[1].plot(range(1, len(ada2.cost_)+1), ada2.cost_, marker='o')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('log(Sum-squared-error)')
ax[1].set_title('Adaline - Learning rate 0.0001')

# show
plt.show()
```
